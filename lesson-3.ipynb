### Exercise 3

#### p.1 - circle

```python
R = 5
x0 = 1
y0 = 1
```

Для окружности удобнее всего использовать следующее представление

```python
alpha = np.linspace(0, 2*np.pi, 3600)
x = R*np.cos(alpha) + x0
y = R*np.sin(alpha)+ y0
```

Построим график

```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = 10, 10
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
```

```python
plt.plot(x, y, color='blue', linewidth=2, alpha=3)
plt.plot(x0, y0, color='blue', marker='.', markersize=16)
plt.axhline(0, color='k')
plt.axvline(0, color='k')
plt.title(f'Circle with radius {R} and center in ({x0}, {y0}) plot')
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.show()
```

#### p.2 - ellipse

```python
a = 8
b = 3
x0 = -0.5
y0 = 0.5
```

Для эллипса можно поступить также

#### p.3 - hyperbola

(x - x0)^2/a - (y - y0)^2/b = 1

```python
x0 = 2
y0 = 3
a = 4
b = 5
```

Для эллипса можно поступить также

```python
x = np.linspace(-2, 6, 10**5)
y1 = -np.sqrt(b*((x - x0)**2/a - 1)) + y0
y2 = np.sqrt(b*((x - x0)**2/a - 1)) + y0
```

```python
plt.plot(x, y1, color='blue', linewidth=2, alpha=3)
plt.plot(x, y2, color='blue', linewidth=2, alpha=3)
plt.axhline(0, color='k')
plt.axvline(0, color='k')
plt.axhline(y0, linestyle='--', color='red', alpha=0.5)
plt.axvline(x0, linestyle='--', color='red', alpha=0.5)
plt.title(f'Hyperbola plot')
plt.xlim(-2, 7)
plt.ylim(-2, 7)
plt.show()
```

### Exercise 5

#### p.1 parallel planes

```python
X = np.linspace(-5, 5, 11)
Y = np.linspace(-5, 5, 11)
X, Y = np.meshgrid(X, Y)
Z = 3*X + 2*Y
Z2 = Z + 3
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z2, color='red')
plt.show()
```
